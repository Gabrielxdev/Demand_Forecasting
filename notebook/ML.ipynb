```python
import pandas as pd 
import numpy as np 

from sklearn.model_selection import train_test_split, RandomizedSearchCV 
from sklearn.preprocessing import LabelEncoder 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBRegressor 
import pickle
```

```python
df = pd.read_csv("../data/demand_forecasting.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
features = [
    "Price", 
    "Discount",
    "Inventory Level",
    "Promotion",
    "Competitor Pricing",
    "Category"
]
```

```python
target = "Demand"
```

```python
X = df[features].copy()
```

```python
y = df[target]
```

```python
X
```

```python
label_encoders = {}

categorical_cols = X.select_dtypes(include="object").columns
```

```python
for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le
```

```python
label_encoders
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```





### 2. `objective = "reg:squarederror"`
*   **O que é:** O "objetivo" define **como o modelo vai medir o próprio erro** durante o treinamento para tentar melhorar.
*   **O que faz:** `reg:squarederror` significa *Regression with Squared Error* (Regressão com Erro Quadrático). Em termos simples, quando o modelo faz uma previsão, ele compara com o valor real, calcula a diferença, eleva essa diferença ao quadrado e tenta minimizar esse valor ao longo do tempo. É a métrica padrão e mais clássica para problemas de regressão.

### 3. `n_jobs = -1`
*   **O que é:** Define a quantidade de núcleos de processamento (cores do seu processador) que serão usados.
*   **O que faz:** O valor `-1` é um truque clássico e muito útil. Ele diz ao computador: *"Use **todos** os núcleos de processamento disponíveis da minha máquina ao mesmo tempo"*. Isso faz com que o treinamento do modelo seja significativamente mais rápido através do processamento paralelo.



```python
xgb = XGBRegressor(objective = "reg:squarederror", n_jobs = -1 )
```

```python
param_dict = {
    "n_estimators": [200, 300, 500],
    "max_depth": [3, 4, 6, 8], 
    "learning_rate": [0.01, 0.05, 0.1], 
    "subsample": [0.7, 0.8, 1.0],
    "colsample_bytree": [0.7, 0.8, 1.0], 
    "min_child_weight": [1,3,5]
}
```

```python
random_search = RandomizedSearchCV(
    estimator = xgb, 
    param_distributions = param_dict, 
    n_iter = 25, 
    scoring = "neg_mean_absolute_error", 
    cv = 3, 
    verbose = 1, 
    n_jobs = -1,
    random_state = 42 # <<-- Garante que o sorteio seja sempre o mesmo nas próximas execuções
)
```

```python
random_search.fit(X_train, y_train)
```

```python
random_search.best_params_
```

```python
best_model = random_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
mse = mean_absolute_error(y_test, y_pred)

rmse = np.sqrt(mse)
print(f"O RMSE do meu modelo é: {rmse}")
```
