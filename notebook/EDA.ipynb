

### 📋 Dicionário de Dados (Features)

*   **Date**: A data em que os dados foram registrados.
*   **Store ID** / **Product ID**: Códigos únicos de identificação da Loja e do Produto, respectivamente.
*   **Category**: A categoria a qual o produto pertence (ex: *Electronics*, *Clothing*, *Groceries*).
*   **Region**: Região geográfica onde a loja está localizada (ex: *North*).
*   **Inventory Level**: A quantidade de peças daquele produto que estava fisicamente disponível no estoque da loja no momento.
*   **Units Sold**: Quantidades do produto que foram efetivamente passaradas pelo caixa (vendas concretizadas).
*   **Units Ordered**: Quantidade de reposição deste produto que a loja encomendou junto aos fornecedores / centro de distribuição.
*   **Price**: O preço adotado pela loja para vender o produto naquele momento.
*   **Discount**: O valor ou percentual de desconto que foi aplicado sobre o preço original no dia.
*   **Weather Condition**: Condição climática daquele dia (ex: *Snowy* / Neve). **Boa prática:** No varejo físico, o clima afeta drasticamente o movimento das pessoas nas ruas e, consequentemente, as vendas.
*   **Promotion**: Indicador (0 = Não, 1 = Sim) se o produto estava no meio de alguma campanha de marketing ou em destaque no dia.
*   **Competitor Pricing**: O preço que a concorrência estava cobrando pelo mesmo produto. Ajuda a medir nosso nível de competitividade.
*   **Seasonality**: A sazonalidade / estação do ano (ex: *Winter* / Inverno).
*   **Epidemic**: Indicador (0 = Não, 1 = Sim) de eventual emergência sanitária que estaria impactando o deslocamento do público.
*   **Demand**: A real intenção de compra do mercado consumidor.

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.day_name()
```

```python
df.head(1)
```

```python
df["Discounted Price"] = df["Price"] * (1 - df["Discount"] / 100)
```

```python
df["Sell Through Rate"] = df["Units Sold"] / df["Inventory Level"]
```
